## Polynomial Regression Example

```python
import numpy as np
from matplotlib.pyplot import *

x = [-7.30000, -4.10000, -1.70000, -0.02564,
     1.50000, 4.50000, 9.10000]
y = [-0.80000, -0.50000, -0.20000, 0.00000,
     0.20000, 0.50000, 0.80000]

coefficients = np.polyfit(x, y, 5)
polynomial = np.poly1d(coefficients)
ys = polynomial(x)
print(coefficients) 
print(polynomial)

plot(x, y, 'o')
plot(x, ys)
ylabel('y')
xlabel('x')
xlim(-10,10)
ylim(-1,1)
show()
```

```python
import numpy as np
from matplotlib.pyplot import *

x = [2.53240, 1.91110, 1.18430, 0.95784, 0.33158,
     -0.19506, -0.82144, -1.64770, -1.87450, -2.2010]

y = [-2.50400, -1.62600, -1.17600, -0.87400, -0.64900,
     -0.477000, -0.33400, -0.20600, -0.10100, -0.00600]

coefficients = np.polyfit(x, y, 2)
polynomial = np.poly1d(coefficients)
xs = np.arange(-2.2, 2.6, 0.1)
ys = polynomial(xs)
print(coefficients)

plot(x, y, 'o')
plot(xs, ys)
ylabel('y')
xlabel('x')
show()
```
